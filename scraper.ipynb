```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36"}
```

```python
books = []
links = []

for i in range(10):
    url = "https://www.kitapyurdu.com/index.php?route=product/category&page="+ str(i+1) +"&filter_category_all=true&path=1_128&filter_in_stock=1&sort=purchased_365&order=DESC&limit=100"
    r = requests.get(url, headers=headers).content
    soup = BeautifulSoup(r, "html.parser")
    all_data = soup.find("div", {"class":"content-bg"})
    
    for book in all_data.find_all("div", {"class":"name ellipsis"}):
        book_name = book.find("span").text.strip()

        if(book_name not in books):
            books.append(book_name)
            links.append(book.find("a").get("href"))
```

```python
ids = [link.split("/")[-1].split(".")[0] for link in links]
```

```python
df = pd.DataFrame({"id":ids, "bookName":books, "link":links})
df.to_csv("linksV2.csv", encoding='utf-16', index=False)
```

```python
df = pd.read_csv("linksV2.csv", encoding='utf-16')
array = df.to_numpy()
```

```python
data = []
for i in range(len(array)):
    try:
        print(array[i][1])
        r = requests.get(array[i][2], headers=headers).content
        soup = BeautifulSoup(r, "html.parser")
        writer = soup.find("div", {"class":"pr_producers__item"}).find("a").contents[0].strip()
        publisher = soup.find("div", {"class":"pr_producers__publisher"}).find("a").contents[0].strip()
        price = soup.find("div", {"class":"price__item"}).contents
        price = price[1].strip()[:-1] + "." + price[2].text
        num_comments = soup.find("div", {"class":"pr_view"}).find_all("span")[1].contents[0].strip()
        num_favorites = soup.find("span", {"class":"favorite-count"}).contents[0].strip()
        num_readers = soup.find("div", {"class":"pr_details__box pr_details__readlist"}).find_all("span", {"class":"readlist-count"})
        num_will_read = num_readers[0].contents[0].strip()
        num_is_read = num_readers[1].contents[0].strip()
        num_did_read = num_readers[2].contents[0].strip()
        num_purchases = soup.find("p", {"class":"purchased"}).contents[0].strip().split(" ")[2]
        num_pages = soup.find("div", {"class":"attributes"}).find_all("td")
        num_pages = num_pages[[num_pages.index(i) for i in num_pages if 'Sayfa' in i.text][0]+1].contents[0]
        type = soup.find("a", {"class":"rel-cats__link"}).find_all("span")[2].contents[0].strip()
        rate_url = "https://www.kitapyurdu.com/index.php?route=product/product/rating&product_id=" + str(array[i][0])
        
        r2 = requests.get(rate_url, headers=headers).content
        soup2 = BeautifulSoup(r2, "html.parser")
        num_ratings = soup2.find("div", {"class":"fl mg-r-10 mg-b-5"}).contents[4].contents[0].strip().split(" ")[0]
        avg_rating = soup2.find("div", {"class":"fl mg-r-10 mg-b-5"}).contents[1].contents[0].strip()
        stars = soup2.find_all("td", {"class":"count"})
        num_five_star = stars[0].text.strip().split(" ")[0]
        num_four_star = stars[1].text.strip().split(" ")[0]
        num_three_star = stars[2].text.strip().split(" ")[0]
        num_two_star = stars[3].text.strip().split(" ")[0]
        num_one_star = stars[4].text.strip().split(" ")[0]

        review_url = "https://www.kitapyurdu.com/index.php?route=product/product/reviewAll&product_id=" + str(array[i][0])
        r3 = requests.get(review_url, headers=headers).content
        soup3 = BeautifulSoup(r3, "html.parser")
        reviews = soup3.find_all("div", {"class":"review"})

        for review in reviews:
            comment = review.find("div", {"class":"review-text"}).contents[0]
            star = len(review.find_all("i", {"class":"sprite__product sprite__product--star-selected"}))
            row = [array[i][0], array[i][1], writer, publisher, num_pages, num_comments, num_favorites, num_will_read, 
            num_is_read, num_did_read, num_purchases, num_ratings, avg_rating, num_five_star, num_four_star,
            num_three_star, num_two_star, num_one_star, comment, price, type, star]
            data.append(row)

    except:
        print(array[i][1], " is passed")
        continue
```
